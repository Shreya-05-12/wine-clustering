```python
for col in df.select_dtypes(include=[np.number]).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df[col] >= (Q1 - 1.5 * IQR)) & (df[col] <= (Q3 + 1.5 * IQR))]
print("\nShape after removing outliers:", df.shape)
```

FEATURE ENGINEERING
    Handling missing values(removing missing data)
    encoding categorical columns
    feature scaling
    creating new features
    reducing dimensionality

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.stats import zscore
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

handling missing values

```python
df.isnull().sum()
```

```python
# Remove outliers using Z-score
z_scores = df_scaled.apply(zscore)
df_cleaned = df_scaled[(z_scores.abs() < 3).all(axis=1)]
```

```python
df_cleaned
```

```python
# Apply PCA (Principal Component Analysis) for dimensionality reduction
# PCA helps in reducing the number of features while retaining maximum variance in the data
pca = PCA(n_components=5)  # Reduce to 5 principal components
df_pca = pd.DataFrame(pca.fit_transform(df_cleaned), columns=[f'PC{i+1}' for i in range(5)])

```

reducing dimensionality(removing less revalent features)

```python
# Compute correlation matrix
corr_matrix = df.corr().abs()

# Create an upper triangle mask
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation > 0.85
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]

print("Features to drop:", to_drop)
```

```python
# Drop highly correlated features
df_reduced = df.drop(columns=to_drop)

print("Reduced dataset shape:", df_reduced.shape)
```
